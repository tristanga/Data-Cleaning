# Snipplet to Slice AirPassengers Data into custom Train / Test using R and xts

```r
options(warn=-1)
library(xts)
library(zoo)
library(lubridate)
data("AirPassengers")
```

```r
df <- as.xts(AirPassengers)
str(df)
```

```r
slice <- 0.7 #slice time series into 70% / 30% 
slice.end <- floor(slice*length(df)) #select the first 70% of the data
train <- df[1:slice.end,] #assign the first 70% of the data to the train set
test <- df[(slice.end+1):length(df),] #assign the most recent 30% to the test set
```

```r
plot(train)
```

Convert back train and test into a Time Serie

```r
ts.test <- ts(as.numeric(test), start = c(year(start(test)), month(start(test))),
                   end = c(year(end(test)), month(end(test))), frequency = 12)
ts.test
plot(ts.test)
```
